```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
csv_path = "/kaggle/input/customer-segmentation-tutorial-in-python/Mall_Customers.csv"

df = pd.read_csv(csv_path)

df.sample(3)
```

```python
df = df.drop(['CustomerID'], axis=1)
```

```python
df = df.rename(columns={
    'Annual Income (k$)': 'Income',
    'Spending Score (1-100)': 'Score'
})
```

```python
df.sample(3)
```

```python
sns.scatterplot(x='Income', y='Score', data=df, hue='Gender', size='Age')
```

```python
sns.histplot(x='Age', data=df, kde=True)
```

```python
sns.histplot(x='Score', data=df, kde=True)
```

```python
r = 3
c = 3

fig, axes = plt.subplots(nrows=r, ncols=c, figsize=(15, 6*r))

nemerical_cols = ['Age', 'Income', 'Score']

axes = axes.flatten()

for i, ax in enumerate(axes):
    nr = i//r
    nc = i%c
    sns.regplot(x=nemerical_cols[nr],y=nemerical_cols[nc], data=df ,ax=ax, line_kws={'color': 'red'})
    # ax.legend(fontsize='small')

plt.tight_layout()
plt.show()
```

```python
r = 3
c = 1

fig, axes = plt.subplots(nrows=r, ncols=c, figsize=(5, 4*r))

nemerical_cols = ['Age', 'Income', 'Score']

axes = axes.flatten()

for i, ax in enumerate(axes):
    nr = i//r
    nc = i%c
    sns.violinplot(x='Gender',y=nemerical_cols[nc], data=df ,ax=ax)
    sns.swarmplot(x='Gender',y=nemerical_cols[nc], data=df ,ax=ax, color='red')
    ax.legend(fontsize='small')

plt.tight_layout()
plt.show()
```

```python
from sklearn.cluster import KMeans

inertia = []

for n in range(1, 11):
    km = KMeans(n_clusters=n)
    km.fit(df[['Age', 'Score']])
    inertia.append(km.inertia_)
```

```python
inertia
```

```python
sns.lineplot(x=range(1, 11), y=inertia, marker='o')
```

```python
km = KMeans(n_clusters=4)
km.fit(df[['Age', 'Score']])
```

```python
centers = km.cluster_centers_
```

```python
df['age_score_label'] = km.labels_
```

```python
x_min = df['Age'].min()-1
x_max = df['Age'].max()+1

y_min = df['Score'].min()-1
y_max = df['Score'].max()+1
```

```python
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1), np.arange(y_min, y_max, 0.1))
```

```python
xx.shape
```

```python
df[['Age', 'Score']].shape
```

```python
region = np.c_[xx.ravel(), yy.ravel()]
```

```python
region.shape
```

```python
z = km.predict(region)
z.shape
```

```python
z
```

```python
z = z.reshape(xx.shape)
```

```python
z.shape
```

```python
# plt.contour(xx, yy, z, alpha=0.5, cmap='plasma')
plt.contourf(xx, yy, z, alpha=0.5, cmap='viridis')
sns.scatterplot(x='Age', y='Score', hue='age_score_label', data=df)
sns.scatterplot(x=centers[:, 0], y=centers[:, 1], color='red', s=100)
```

```python
km = KMeans(n_clusters=5)
km.fit(df[['Income', 'Score']])
centroids = km.cluster_centers_
```

```python
labels = km.labels_
df['income_score_label'] = labels
```

```python
x_min, x_max = df['Income'].min()-1, df['Income'].max()+1
y_min, y_max = df['Score'].min()-1, df['Score'].max()+1
```

```python
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1), np.arange(y_min, y_max, 0.1))
```

```python
z = km.predict(np.c_[xx.ravel(), yy.ravel()])
z = z.reshape(xx.shape)
```

```python
z
```

```python
plt.contourf(xx, yy, z, cmap='viridis', alpha=0.3)
sns.scatterplot(x='Income', y='Score', hue='income_score_label', data=df)
sns.scatterplot(x=centroids[:, 0], y=centroids[:, 1], s=100, color='red')
```
